Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Function to extract top 10 reviews from IMDb link
def extract_reviews(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        reviews = soup.find_all('div', class_='review-container')[:10] # Extracting top 10 reviews
        return reviews
    else:
        print("Failed to retrieve data from IMDb.")
        return None

# Function to extract sentiment keywords using TF-IDF method
def extract_sentiment_keywords(reviews):
    corpus = [review.find('div', class_='text show-more__control').text.strip() for review in reviews]
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = vectorizer.get_feature_names_out()
    sentiment_keywords = []
    for i, review_text in enumerate(corpus):
        # Extracting sentiment keywords with TF-IDF values greater than a threshold
        keywords = [feature_names[j] for j in tfidf_matrix[i].indices if tfidf_matrix[i,j] > 0.2]
        sentiment_keywords.append(keywords)
    return sentiment_keywords

# IMDb movie review links and corresponding movie names
movie_links = [
    "https://www.imdb.com/title/tt23849204/reviews/?ref_=tt_ov_rt",
    "https://www.imdb.com/title/tt15398776/reviews/?ref_=tt_ov_rt",
    "https://www.imdb.com/title/tt5074352/reviews/?ref_=tt_ov_rt"
]
movie_names = [
    "Godzilla Minus One",
    "Oppenheimer",
    "Sample Movie Name"
]

# Empty list to store extracted data
extracted_data = []

# Extracting reviews, sentiment keywords, and populating DataFrame
for movie_name, link in zip(movie_names, movie_links):
    reviews = extract_reviews(link)
    if reviews:
        for review in reviews:
            # Extracting review date, review text, reviewer username, review score, and sentiment keywords
            review_date = review.find('span', class_='review-date').text.strip()
            review_text = review.find('div', class_='text show-more__control').text.strip()
            reviewer_username = review.find('span', class_='display-name-link').text.strip()
            review_score = review.find('span', class_='rating-other-user-rating').text.strip()
            sentiment_keywords = extract_sentiment_keywords([review])  # Extracting sentiment keywords for the current review

            # Storing extracted data in a list
            extracted_data.append([movie_name, review_date, review_text, reviewer_username, review_score, sentiment_keywords])

# Creating DataFrame from extracted data
dataframe = pd.DataFrame(extracted_data, columns=["Movie Title", "Review Date", "Review Text", "Reviewer Username", "Review Score", "Sentiment Keywords"])

# Displaying the DataFrame
print(dataframe)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Function to extract top 10 reviews from IMDb link
def Movie_reviews(movie_url):
    Web_response = requests.get(movie_url)
    if Web_response.status_code == 200:
        soup_Beauty = BeautifulSoup(Web_response.text, 'html.parser')
        Extracted_reviews = soup_Beauty.find_all('div', class_='review-container')[:10] # Extracting top 10 reviews
        return Extracted_reviews
    else:
        print("Failed to retrieve data from IMDb.")
        return None

# Function to extract sentiment keywords using TF-IDF method
def extract_sentiment_keywords(Extracted_reviews):
    raw_corpus = [review.find('div', class_='text show-more__control').text.strip() for review in Extracted_reviews]
    Tfvectorizer = TfidfVectorizer(stop_words='english')
    matrix_tfidf = Tfvectorizer.fit_transform(raw_corpus)
    extractedFeature_names = Tfvectorizer.get_feature_names_out()
    sentiment_keywords = []
    for i, review_text in enumerate(raw_corpus):
        # Extracting sentiment keywords with TF-IDF values greater than a threshold
        keywords = [extractedFeature_names[j] for j in matrix_tfidf[i].indices if matrix_tfidf[i,j] > 0.2]
        sentiment_keywords.append(keywords)
    return sentiment_keywords

# IMDb movie review links and corresponding movie names
movie_urls = [
    "https://www.imdb.com/title/tt23849204/reviews/?ref_=tt_ov_rt",
    "https://www.imdb.com/title/tt15398776/reviews/?ref_=tt_ov_rt",
    "https://www.imdb.com/title/tt5074352/reviews/?ref_=tt_ov_rt"
]
movie_names = [
    "Godzilla Minus One",
    "Oppenheimer",
    "Sample Movie Name"
]

# Empty list to store extracted data
extracted_raw_data = []

# Extracting reviews, sentiment keywords, and populating DataFrame
for movie_name, url in zip(movie_names, movie_urls):
    Extracted_reviews = Movie_reviews(url)
    if Extracted_reviews:
        for movie_review in Extracted_reviews:
            # Extracting review date, review text, reviewer username, review score, and sentiment keywords
            moviereview_date = movie_review.find('span', class_='review-date').text.strip()
            moviereview_text = movie_review.find('div', class_='text show-more__control').text.strip()
            moviereviewer_username = movie_review.find('span', class_='display-name-link').text.strip()
            moviereview_score = movie_review.find('span', class_='rating-other-user-rating').text.strip()
            sentiment_keywords = extract_sentiment_keywords([movie_review])  # Extracting sentiment keywords for the current review

            # Storing extracted data in a list
            extracted_raw_data.append([movie_name, moviereview_date, moviereview_text, moviereviewer_username, moviereview_score, sentiment_keywords])

# Creating DataFrame from extracted data
dataframe = pd.DataFrame(extracted_raw_data, columns=["Movie Title", "Review Date", "Review Text", "Reviewer Username", "Review Score", "Sentiment Keywords"])

# Displaying the DataFrame
print(dataframe)
```

```python
# Feature selection using SelectKBest
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(dataframe['Review Text'])
y = dataframe['Review Score']  # Assuming review score is the target variable

selector = SelectKBest(chi2, k=1000)  # Select top 1000 features
X_selected = selector.fit_transform(X, y)

# Transforming back to feature names
feature_names = vectorizer.get_feature_names_out()
selected_feature_names = [feature_names[i] for i in selector.get_support(indices=True)]

# Displaying the selected features
print("Selected Features:")
print(selected_feature_names)

# Creating DataFrame to show feature scores
rank_df = pd.DataFrame(selector.scores_, columns=["Scores"], index=feature_names)
rank_df.sort_values(by='Scores', ascending=False, inplace=True)
print(rank_df.head())
```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Load pre-trained BERT model and tokenizer
model_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model_bert = BertModel.from_pretrained('bert-base-uncased')

# Query
input_query = "Its an amazing movie"

# Tokenize query
tokens_query = model_tokenizer.encode(input_query, add_special_tokens=True, return_tensors="pt")

# Generate embeddings for query
with torch.no_grad():
    outputs_query = model_bert(tokens_query)[0][:, 0, :].squeeze()

# Calculate cosine similarity for each review in the text data
reviews_rank = []
similarities_rank = []
for review in dataframe['Review Text']:
    # Truncate review to fit within the maximum sequence length
    review = review[:512]

    # Tokenize review
    review_tokens = model_tokenizer.encode(review, add_special_tokens=True, return_tensors="pt")

    # Generate embeddings for review
    with torch.no_grad():
        review_output = model_bert(review_tokens)[0][:, 0, :].squeeze()

    # Calculate cosine similarity
    similarity = cosine_similarity(outputs_query.reshape(1, -1), review_output.reshape(1, -1))[0][0]

    # Append similarity and review
    similarities_rank.append(similarity)
    reviews_rank.append(review)

# Rank similarities in descending order
ranked_indices = sorted(range(len(similarities_rank)), key=lambda i: similarities_rank[i], reverse=True)
reviews_rank = [reviews_rank[i] for i in ranked_indices]
similarities_rank = [similarities_rank[i] for i in ranked_indices]

# Print ranked text data and similarities
for i, (review, similarity) in enumerate(zip(reviews_rank, similarities_rank), 1):
    print(f"Rank {i}:")
    print(f"Review: {review}")
    print(f"Similarity: {similarity}")
    print()
```
